## Imports

```
import numpy as np
from IPython.display import display, Audio
import time as time

# the functions used to generate audio belong to tensorflow v1,
# so we need to disable the v2 functions.
import tensorflow as tf
tf.compat.v1.disable_v2_behavior()
```

## Model Files
You need to provide the path to the infer.meta file and the model ckpt files (`.data` and `.index`).

```
infer_meta = "/path/to/file/infer.meta"
model_ckpt = "/path/to/file/model.ckpt"
```

### Load the model files into tensorflow

```
# Load the model
tf.compat.v1.reset_default_graph()
saver = tf.compat.v1.train.import_meta_graph(infer_meta)
graph = tf.compat.v1.get_default_graph()
sess = tf.compat.v1.InteractiveSession()
saver.restore(sess, model_ckpt)
```

```
is_specgan = True
```

### Generate audio

#### Number of samples to generate

```
num_generate = 6
```

#### Sample latent vectors

```
_z = (np.random.rand(num_generate, 100) * 2) - 1
```

```
_z.shape
```

#### Fetch tensors

```
z = graph.get_tensor_by_name('z:0')
G_z = graph.get_tensor_by_name('G_z:0')[:, :, 0]

feeds = {}
feeds[z] = _z
if is_specgan:
    ngl = graph.get_tensor_by_name('ngl:0')
    feeds[ngl] = 16 # default value, change if you ran different argument
```

#### Run session to generate audio

```
start = time.time()
_G_z = sess.run(G_z, feeds)
print('Finished! (Took {} seconds)'.format(time.time() - start))
```

#### Generate Audio

```
for i in range(num_generate):
  print('-' * 80)
  print('Example {}'.format(i))
  display(Audio(_G_z[i], rate=16384))
```

#### Close session

```
sess.close()
```
